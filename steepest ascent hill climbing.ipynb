```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mutation(x):
    x += np.random.uniform(low=-2.0, high=2.0)
    return x
```

```python
# Функция, у которой ищется минимум
func = lambda x: x ** 4 + 5 * x ** 3 + 4 * x ** 2 - 4 * x + 1

iteration = 0
# Генерация случайного вещественного числа их промежутка low ... high
x = np.random.uniform(low=-4, high=4)

while iteration != 1000:
    # Мутация x
    x_mutation = mutation(x)
    f = func(x_mutation)
    x = x_mutation if func(x_mutation) < func(x) else x
    iteration += 1
else:
    print("Value function:", func(x), "x:", x)

# Вывод расположенного значения на графике
plt.scatter(x, func(x), c="red")
x = np.linspace(-4, 1.5, num=100)
plt.plot(x, func(x))
plt.show()
```
